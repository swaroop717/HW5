```python
import math                                         
import itertools
f = open('rosalind_perm (1).txt')
n = int(f.readline().strip())                              
print(math.factorial(n))                            
perm = itertools.permutations(list(range(1, n + 1)))
for i, j in enumerate(list(perm)):                  
    l = ''                                
    for k in j:                                  
        l += str(k) + ' '              
    print(l)                              
```
